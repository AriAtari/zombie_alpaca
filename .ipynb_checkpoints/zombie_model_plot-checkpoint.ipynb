# Plotting 

Here, we analyze different behaviors as we change their parameters and measure their impact on the populations of humans and zombies

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import display, clear_output

# Importing necessary classes to run simulation
from zombie_model import human, zombie, alpacalypse 

np.random.seed(123456789) # added so the change in parameters has a clearer impact on the model
```

## Basic Simulation: Changing the population

Here, run the simulation at different starting populations and plot out the number of humans that are infected as a function of the iteration

```python
hpops = [350,300,250,200,150,100,50] # list of human populations

plt.figure(figsize = (15,10))

for num in range(len(hpops)):
    
    my_alpaca = alpacalypse(width = 200 , height = 200)
    
    for p in range(hpops[num]):
        my_alpaca.humans.append(
            human(base_chance_of_survival = 0)
        )
    for p in range(400-hpops[num]):
        my_alpaca.zombies.append(
            zombie(smell_radius = 0)
        )

    my_alpaca._all_entities = [*my_alpaca.humans, *my_alpaca.zombies]

    for hum in my_alpaca.humans:
        hum.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])

    for zom in my_alpaca.zombies:
        zom.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
    
    for i in range(100):
        my_alpaca.evolve()
    
    humans_eaten = hpops[num] - np.array(my_alpaca.human_count)
    
    plt.plot(range(100),humans_eaten,label = "Human Pop = {}".format(hpops[num]))
    plt.xlabel('Time')
    plt.ylabel("Humans Infected")
    plt.legend()
```

## Increasing Zombie Scent Radius

Changing the radius at which a zombie can detect a human to see how it impacts the respecive populations 

```python
zombie_pop = []
human_pop = []
scent_radii = []

i = 0
while i < 24:
    scent_radii.append(i)
    i+=1

i = 24
while i < 51:
    scent_radii.append(i)
    i+=1
    

# Initializing a bunch of humans and zombies
for i in range(len(scent_radii)):
    my_alpaca = alpacalypse(width = 200 , height = 200) # initializing apocolypse class
    
    for p in range(200):
        my_alpaca.humans.append(
            human(base_chance_of_survival = 0)
        )

        my_alpaca.zombies.append(
            zombie(smell_radius = scent_radii[i]) # setting smell radius to 0 to isolate desired features
        )

    # Group both sets of entities in one list
    my_alpaca._all_entities = [*my_alpaca.humans, *my_alpaca.zombies]

    print(i,"Done")

    for hum in my_alpaca.humans:
        hum.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
    
    for zom in my_alpaca.zombies:
        zom.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
        
        
    for i in range(100):
        # RUN YOUR ABM
        my_alpaca.evolve()
        
        
    zombie_pop.append(len(my_alpaca.zombies))
    human_pop.append(len(my_alpaca.humans))
```

```python
plt.figure(figsize=(10,8))
plt.plot(scent_radii,zombie_pop,color="red",label="Zombies")
plt.plot(scent_radii,human_pop,color="blue",label="Humans")
plt.legend()
plt.ylabel("Population Count")
plt.xlabel("Scent Radius")
plt.title("Final Populations as a Function of Smell Radius")
```

## Changing Human Survival Chance

Changing the base human survival chance to see how it impacts the human population

```python
np.random.seed(123456789)

chance = [1,2,3,4,5]

fig,ax = plt.subplots(1,1, figsize = (16,9))

for ch in chance:
    
    my_alpaca = alpacalypse(width = 200 , height = 200)
    
    for p in range(200):
        hum = human(base_chance_of_survival = ch)
        
        my_alpaca.humans.append(hum)
    for p in range(200):
        my_alpaca.zombies.append(
            zombie(smell_radius = 0)
        )

    my_alpaca._all_entities = [*my_alpaca.humans, *my_alpaca.zombies]

    for hum in my_alpaca.humans:
        hum.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])

    for zom in my_alpaca.zombies:
        zom.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
    
    for i in range(100):
        my_alpaca.evolve()
    
    human_numbers = np.array(my_alpaca.human_count)/200
    zombie_numbers = np.array(my_alpaca.zombie_count)/200
    
    ax.plot(range(100),human_numbers,label=str(ch))
    plt.xlabel("iterations")
    plt.ylabel("percentage of humans left")
    plt.legend()
```

## Let Them Eat Lead

Giving humans different weapons to view how it impacts their survivability

```python
np.random.seed(123456789)

weapon = ["none","pistol","rifle","grenade","rocket launcher"]

fig,ax = plt.subplots(1,1, figsize = (16,9))

for wp in weapon:
    
    my_alpaca = alpacalypse(width = 200 , height = 200)
    
    for p in range(200):
        hum = human(base_chance_of_survival = 1) # can't let it be 0 otherwise weapons provide no benefit
        hum.gets_weapon(wp) # provdes human with a weapon to increase their survival chance
        my_alpaca.humans.append(hum)
    for p in range(200):
        my_alpaca.zombies.append(
            zombie(smell_radius = 0)
        )

    my_alpaca._all_entities = [*my_alpaca.humans, *my_alpaca.zombies]

    for hum in my_alpaca.humans:
        hum.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])

    for zom in my_alpaca.zombies:
        zom.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
    
    for i in range(100):
        my_alpaca.evolve()
    
    human_numbers = np.array(my_alpaca.human_count)/200
    zombie_numbers = np.array(my_alpaca.zombie_count)/200
    
    ax.plot(range(100),human_numbers,label=str(wp))
    plt.xlabel("iterations")
    plt.ylabel("percentage of humans left")
    plt.legend()
```
