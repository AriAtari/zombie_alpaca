### Basic simulation

```python
# Importing necessary libraries to run simulation
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import display, clear_output

# Importing necessary classes to run simulation
from zombie_model import human, zombie, alpacalypse 
```

The following is a basic simulation that simply adapts our 'standard_model.py' code into a zombie apocalypse model. There are no real special behaviors in the humans or the zombies. They merely randomly wander around at similar speeds and if they interact one another, the human becomes a zombie.

Check 'zombie_model.py' for more specifics on how each of the functions work and how one might add other behaviors to the model

```python
np.random.seed(123456789) # setting a seed so that analysis is repreatable

my_alpaca_base = alpacalypse(width = 200 , height = 200) # initializing alpacalypse class

# Initializing a bunch of humans and zombies
for p in range(200):
    my_alpaca_base.humans.append(
        human(base_chance_of_survival = 0) # setting to 0 to remove complex behaviors
    )
    
    my_alpaca_base.zombies.append(
        zombie(smell_radius = 0) # setting to 0 to remove complex behaviors
    )

# Group both sets of entities in one list
my_alpaca_base._all_entities = [*my_alpaca_base.humans, *my_alpaca_base.zombies]
```

```python
# Testing to see if the '.place_at' and '.draw()'functions work
for hum in my_alpaca_base.humans:
    hum.place_at([np.random.randint(low = 0, high = my_alpaca_base.width),np.random.randint(low = 0, high = my_alpaca_base.height)])
    
for zom in my_alpaca_base.zombies:
    zom.place_at([np.random.randint(low = 0, high = my_alpaca_base.width),np.random.randint(low = 0, high = my_alpaca_base.height)])

my_alpaca_base.draw()
```

```python
# Using the animation code from HW2

fig, ax = plt.subplots(figsize = (10,10))

start = time.time()
times = []

for i in range(100):
    
    # Running the ABM
    my_alpaca_base.evolve()
    my_alpaca_base.draw()
    
    # same animation code we used in hw2
    plt.title(f"Time = {i}")
    clear_output(wait=True) 
    display(fig)            
    fig.clear()             
        
    end = time.time()
    times.append(end-start)
    
print("Total time =", end-start)    
```

```python
# Sometimes I get a bit of an odd discrepancy between the number of infected zombies and the number of created zombies
print("The number of humans to be turned into zombies was",200 - len(my_alpaca_base.humans))
print("The number of zombies that were created was",len(my_alpaca_base.zombies)-200)
```

### Ability to smell humans
#### The zombies have better path finding that allows them to detect humans within a certain radius and move towards them

```python
%reload_ext autoreload
%autoreload 2
%pylab
%matplotlib inline

from IPython.display import display, clear_output

from zombie_model import human, zombie, alpacalypse # importing necessary classes

my_alpaca_smell = alpacalypse(width = 200 , height = 200) # initializing apocolypse class

np.random.seed(123456789)

# Initializing a bunch of humans and zombies
for p in range(200):
    my_alpaca_smell.humans.append(
        human(base_chance_of_survival = 0) # setting smell radius to -1 to 
                                            # isolate desired features (zombie scent)
    )

    my_alpaca_smell.zombies.append(
        zombie(smell_radius = 10)
    )

# Group both sets of entities in one list
my_alpaca_smell._all_entities = [*my_alpaca_smell.humans, *my_alpaca_smell.zombies]

print("Done")

for hum in my_alpaca_smell.humans:
    hum.place_at([np.random.randint(low = 0, high = my_alpaca_smell.width),np.random.randint(low = 0, high = my_alpaca_smell.height)])
    
for zom in my_alpaca_smell.zombies:
    zom.place_at([np.random.randint(low = 0, high = my_alpaca_smell.width),np.random.randint(low = 0, high = my_alpaca_smell.height)])
```

```python
fig, ax = plt.subplots(figsize = (10,10))

start = time.time()
times = []

for i in range(100):
    
    # Running the ABM
    my_alpaca_smell.evolve()
    my_alpaca_smell.draw()
    
    # same animation code we used in hw2
    plt.title(f"Time = {i}")
    clear_output(wait=True) 
    display(fig)            
    fig.clear()             
        
    end = time.time()
    times.append(end-start)
    
print("Total time =", end-start)
```

```python
print("The number of humans to be turned into zombies was",200 - len(my_alpaca_smell.humans))
print("The total number of zombies changed by",len(my_alpaca_smell.zombies)-200)
```

Here, in addition to being able to see zombies follow humans in the animaation, we see that more people were infected even with the same seed, implying that our scent behavor is functioning as it should

### Ability to fight back and tain
#### Humans can start out with a set chance of surviving a zombie attack by killing the zombie
#### After each zombie attack a human will improve their chances of survival

```python
%reload_ext autoreload
%autoreload 2
%pylab
%matplotlib inline

my_alpaca_fight = alpacalypse(width = 200 , height = 200) # initializing apocolypse class

np.random.seed(123456789)

# Initializing a bunch of humans and zombies
for p in range(200):
    my_alpaca_fight.humans.append(
        human(base_chance_of_survival = 1)
    )

    my_alpaca_fight.zombies.append(
        zombie(smell_radius = 0) # setting smell radius to 0 to isolate desired features
    )

# Group both sets of entities in one list
my_alpaca_fight._all_entities = [*my_alpaca_fight.humans, *my_alpaca_fight.zombies]

print("Done")

for hum in my_alpaca_fight.humans:
    hum.place_at([np.random.randint(low = 0, high = my_alpaca_fight.width),np.random.randint(low = 0, high = my_alpaca_fight.height)])
    
for zom in my_alpaca_fight.zombies:
    zom.place_at([np.random.randint(low = 0, high = my_alpaca_fight.width),np.random.randint(low = 0, high = my_alpaca_fight.height)])
```

```python
fig, ax = plt.subplots(figsize = (10,10))

start = time.time()
times = []

for i in range(100):
    
    # Running the ABM
    my_alpaca_fight.evolve()
    my_alpaca_fight.draw()
    
    # same animation code we used in hw2
    plt.title(f"Time = {i}")
    clear_output(wait=True) 
    display(fig)            
    fig.clear()             
        
    end = time.time()
    times.append(end-start)
    
print("Total time =", end-start)
```

```python
print("The number of humans to be turned into zombies was",200 - len(my_alpaca_fight.humans))
print("The total number of zombies changed by",len(my_alpaca_fight.zombies)-200)
```

Here we see, from the initial population, the number of zombies decrese by 3, showing that the humans are, in fact, capable of fighting back against the zombies

### Getting weapons
#### Provide humans with weapons that increase the chance of success

```python
%reload_ext autoreload
%autoreload 2
%pylab
%matplotlib inline

my_alpaca_weapon = alpacalypse(width = 200 , height = 200) # initializing apocolypse class

np.random.seed(123456789)

# Initializing a bunch of humans and zombies
for p in range(200):
    hum = human(base_chance_of_survival = 1) # had to slightly alter how human objects were created to give them weapons
    hum.gets_weapon("pistol") # running the gets_weapon() function to give each human a pistol
                              # by changing "pistol" to any of the 4 options we see different survival 
                              # rates between the humans
        
    my_alpaca_weapon.humans.append(hum)

    my_alpaca_weapon.zombies.append(
        zombie(smell_radius = 0) # setting smell radius to 0 to isolate desired features
    )

# Group both sets of entities in one list
my_alpaca_weapon._all_entities = [*my_alpaca_weapon.humans, *my_alpaca_weapon.zombies]

print("Done")

for hum in my_alpaca_weapon.humans:
    hum.place_at([np.random.randint(low = 0, high = my_alpaca_weapon.width),np.random.randint(low = 0, high = my_alpaca_weapon.height)])
    
for zom in my_alpaca_weapon.zombies:
    zom.place_at([np.random.randint(low = 0, high = my_alpaca_weapon.width),np.random.randint(low = 0, high = my_alpaca_weapon.height)])
```

```python
fig, ax = plt.subplots(figsize = (10,10))

start = time.time()
times = []

for i in range(100):
    
    # Running the ABM
    my_alpaca_weapon.evolve()
    my_alpaca_weapon.draw()
    
    # same animation code we used in hw2
    plt.title(f"Time = {i}")
    clear_output(wait=True) 
    display(fig)            
    fig.clear()             
        
    end = time.time()
    times.append(end-start)
    
print("Total time =", end-start)
```

```python
print("The number of humans to be turned into zombies was",200 - len(my_alpaca_weapon.humans))
print("The total number of zombies changed by",len(my_alpaca_weapon.zombies)-200)
```

Here, we see that the humans were able to kill more zombies and with the base survival rate of 1, showing that the weapons behaviors are functioning properly

### Plotting the implements

Most of the code is located in zombie_model_plot.ipynb. The following shows an example of how one would plot out the human and zombie populations after each iteration

```python
# Using base simulation code

my_alpaca_base = alpacalypse(width = 200 , height = 200) # initializing alpacalypse class

# Initializing a bunch of humans and zombies
for p in range(200):
    my_alpaca_base.humans.append(
        human(base_chance_of_survival = 0) # setting to 0 to remove complex behaviors
    )
    
    my_alpaca_base.zombies.append(
        zombie(smell_radius = 0) # setting to 0 to remove complex behaviors
    )

# Group both sets of entities in one list
my_alpaca_base._all_entities = [*my_alpaca_base.humans, *my_alpaca_base.zombies]

for hum in my_alpaca_base.humans:
    hum.place_at([np.random.randint(low = 0, high = my_alpaca_base.width),np.random.randint(low = 0, high = my_alpaca_base.height)])
    
for zom in my_alpaca_base.zombies:
    zom.place_at([np.random.randint(low = 0, high = my_alpaca_base.width),np.random.randint(low = 0, high = my_alpaca_base.height)])

for i in range(100):
    # Running the ABM without plotting
    my_alpaca_base.evolve()
```

```python
# Lists containing the number of humans and zombies at the end of each timestep
human_numbers = my_alpaca_base.human_count
zombie_numbers = my_alpaca_base.zombie_count

print(len(human_numbers))
```

```python
# Here's an example of how one could make a plot of the human and zombie counts over time

plt.plot(range(100),human_numbers, label = "Human population")
plt.plot(range(100),zombie_numbers, label = "Zombie population")
plt.xlabel("Time (iterations)")
plt.ylabel("Population")
plt.legend()
```
