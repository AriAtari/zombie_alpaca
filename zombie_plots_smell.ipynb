## Increasing Scent Radius

```python
import numpy as np
import matplotlib.pyplot as plt
import time
from IPython.display import display, clear_output

# Importing necessary classes to run simulation
from zombie_model import human, zombie, alpacalypse 
```

```python
np.random.seed(123456789)

zombie_pop = []
human_pop = []
scent_radii = []

i = 0
while i < 24:
    scent_radii.append(i)
    i+=1

i = 24
while i < 51:
    scent_radii.append(i)
    i+=1
    

# Initializing a bunch of humans and zombies
for i in range(len(scent_radii)):
    my_alpaca = alpacalypse(width = 200 , height = 200) # initializing apocolypse class
    
    for p in range(200):
        my_alpaca.humans.append(
            human(base_chance_of_survival = 0)
        )

        my_alpaca.zombies.append(
            zombie(smell_radius = scent_radii[i]) # setting smell radius to 0 to isolate desired features
        )

    # Group both sets of entities in one list
    my_alpaca._all_entities = [*my_alpaca.humans, *my_alpaca.zombies]

    print(i,"Done")

    for hum in my_alpaca.humans:
        hum.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
    
    for zom in my_alpaca.zombies:
        zom.place_at([np.random.randint(low = 0, high = my_alpaca.width),np.random.randint(low = 0, high = my_alpaca.height)])
        
        
    for i in range(100):
        # RUN YOUR ABM
        my_alpaca.evolve()
        
        
    zombie_pop.append(len(my_alpaca.zombies))
    human_pop.append(len(my_alpaca.humans))
        
```

```python
plt.figure(figsize=(10,8))
plt.plot(scent_radii,zombie_pop,color="red",label="Zombies")
plt.plot(scent_radii,human_pop,color="blue",label="Humans")
plt.legend()
plt.ylabel("Population Count")
plt.xlabel("Scent Radius")
plt.title("Final Populations as a Function of Smell Radius")
```
